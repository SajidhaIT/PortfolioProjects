# Exploratory Data Analysis (EDA) in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\to_no\Downloads\world_population01.csv')
```

```python
df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
df.corr(numeric_only=True)  #Correlation
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['float'])
sns.heatmap(numeric_columns.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (20, 7)
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df[df.Continent=='Oceania']
```

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending = False)
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending = False)

df2.plot()
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
df.columns
```

```python
df4 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population', ascending = False)
df4
```

```python
df5 = df.groupby('Continent')[['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending = False)
df5
```

```python
df5.plot()
```

```python
df5 = df5.transpose()
df5
```

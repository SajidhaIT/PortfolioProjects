# Creating Visualizations using Pandas Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\to_no\Downloads\Ice Cream Ratings.csv')
df
```

```python
df.set_index('Date')
df
```

```python
df.plot()
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', subplots = True, title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar',  title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar',  title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar',  title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh( title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```

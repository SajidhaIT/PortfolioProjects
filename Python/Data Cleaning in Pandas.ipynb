# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\to_no\Downloads\Customer Call List.xlsx')
df
```

```python
# Remove Duplicates
df = df.drop_duplicates()
df
```

```python
# Remove unwanted columns
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("...")
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("/")
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('_')
df
```

```python
# SECOND METHOD
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
# Replace
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[7:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
df['Address']
```

```python

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(",", n = 2, expand = True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
```

```python
# Replace NaN, None to empty
df = df.fillna('')
df
```

```python
# Get Data with phone numbers and who have not applied Do Not Contact

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)

df
    
```

```python


df = df.dropna(subset = 'Phone Number', inplace = True)
df
```

```python
# Reset the index
df = df.reset_index(drop=True)
```

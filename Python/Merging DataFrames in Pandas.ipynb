# Merge, Join and Concatenate DataFrames in Pandas

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'C:\Users\to_no\Downloads\LOTR.csv')
df1
```

```python
df2 = pd.read_csv(r'C:\Users\to_no\Downloads\LOTR 2.csv')
df2
```

# Merge

```python
# Default merge is inner join
df1.merge(df2)  
```

```python
# Inner Join on FellowshipID

df1.merge(df2, how = 'inner', on = 'FellowshipID')  
```

```python
# Inner Join on FellowshipID & FirstName

df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])  
```

```python
# Outer Join 

df1.merge(df2, how = 'outer')  
```

```python
# Left Join 

df1.merge(df2, how = 'left')  
```

```python
# Right Join 

df1.merge(df2, how = 'right')  
```

```python
# cross Join 

df1.merge(df2, how = 'cross')  
```

# Joins

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
#Joins with index
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
#Outer join with index
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

# Concatenate


```python
pd.concat([df1,df2])
```

```python
#Joins using concat

pd.concat([df1,df2], join = 'inner')
```

```python
#Joins using concat

pd.concat([df1,df2], join = 'outer')
```

```python
#Joins using concat

pd.concat([df1,df2], join = 'outer', axis = 1)
```
